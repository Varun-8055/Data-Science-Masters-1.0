### DBSCAN Clustering Anomaly Detection

```python
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
%matplotlib inline 
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)#we are adding some noise otherwise all datapoints will be in 1 line  
```

```python
# Here we are getting the y features as well but we will not use this as we are performing unsupervised learning
```

```python
X    # we have 2 features 
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbscan=DBSCAN(eps=0.10)
```

```python
dbscan.fit_predict(X) # -1 is outliers
```

```python
dbscan.labels_
```

```python
 plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)  # dark voilet is outliers 
```

### Local Outlier Anomaly Detection

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof=LocalOutlierFactor(n_neighbors=5,novelty=True)
```

```python
lof.fit(X)
```

```python
predictions=lof.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=predictions)  # voilet ones are the local outliers
```
