### DBSCAN Implementation
DBSCAN helps in cluster non linear separable points

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons  # make two interleaving half circle (non-linear datasets)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_moons(n_samples=250,noise=0.15)
```

```python
# X  # just use X features
```

```python
plt.scatter(X[:,0],X[:,1])  # non linear datasets 
```

```python
# feature scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
# X_scaled
```

```python
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_  # 0 and 1 label is fine but the -1 label is the outliers. 2 label is border points
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)  # the voilets ones are outliers
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    dbscan=DBSCAN(eps=0.3)
    dbscan.fit(X_scaled)
    score=silhouette_score(X_scaled,dbscan.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xlabel('No. of clusters')
plt.ylabel('silhouette_coefficients')
plt.show()
```
