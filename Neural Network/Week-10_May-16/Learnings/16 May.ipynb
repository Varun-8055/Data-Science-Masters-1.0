```python
import numpy as np 
import matplotlib.pyplot as plt 
import joblib 
import pandas as pd
```

```python
class perceptron:
    # weight get initialized here
    def __init__(self,eta,epochs):
        self.weights=np.random.randn(3)* 1e-4
        print(f"Weights are:{self.weights}")
        self.eta=eta
        self.epochs=epochs
       
    # here w1*x1 + w2*x2 + w3*x3 + ---- + Wn*xn will be calculated 
    def activation_function(self,inputs,weights):
        z=np.dot(inputs,weights)
        return np.where(z>0,1,0) # weights will get updated according to the threshold we sets 
    
    # here weights get updated Wnew = Wold - (alpha * xi * [partial diff of loss / partial diff of weights])
    def fit(self,X,y):
        self.X=X
        self.y=y
        
        X_with_bias=np.c_[self.X,-np.ones((len(self.X),1))]                                   # dot product 
        print(f"X_with_bias \n {X_with_bias}") # w1*x1+w2*x2+w3*x3+b can be written as [x1 x2 x3 1].[w1 w2 w3 b]
        
        for epochs in range(self.epochs):
            print(f"for epoch: {epochs}")
            y_hat=self.activation_function(X_with_bias,self.weights)
            print(f"predicted value: \n{y_hat}")
            error=self.y-y_hat
            print(f"Error is: \n{error}")
            self.weights=self.weights+self.eta*np.dot(X_with_bias.T,error)
            print(f"Updated weights is: {self.weights}")
            
    def predict(self,X):
        X_with_bias=np.c_[X,-np.ones((len(self.X),1))]   # concatenate
        return self.activation_function(X_with_bias,self.weights)
```

### AND Operation

```python
data={"x1":[0,0,1,1],"x2":[0,1,0,1],"y":[0,0,0,1]}
```

```python
AND=pd.DataFrame(data)
AND
```

```python
X=AND.drop('y',axis=1)
```

```python
X
```

```python
y=AND['y']
y.to_frame()
```

```python
model=perceptron(eta=0.5,epochs=10)
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.weights
```
